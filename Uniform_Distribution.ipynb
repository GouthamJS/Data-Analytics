```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#creating data frame dor uniform distribution
#loc->lower bound
#scale->upper bound
uniform_data=stats.uniform.rvs(size=1000,loc=0,scale=10)
```

```python
uniform_data
```

```python
pd.DataFrame(uniform_data).plot(kind="density",figsize=(6,6),xlim=(-1,11))
```

```python
#density plot
#pdf
for x in range(-1,12,3):
    print("Density at x value"+str(x))
    print(stats.uniform.pdf(x,loc=0,scale=10))
```
