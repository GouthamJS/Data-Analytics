```python
#Visualization of Binomial Distribution
#kde - kernal density estimation is aapplication of kernal smoothing
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.binomial(n=10,p=0.5,size=1000),hist=True,kde=False)
plt.show()
```

```python
sns.distplot(random.binomial(n=10,p=0.5,size=1000),hist=True,kde=True)
plt.show()
```

```python
#Normal distribution curve
sns.distplot(random.normal(loc=50,scale=5,size=1000),hist=True,kde=True)
plt.show()
```

```python
sns.distplot(random.normal(loc=50,scale=5,size=1000),hist=False,kde=True)
plt.show()
```

```python
import scipy.stats as stats
#cdf - cumulative distribution function
stats.uniform.cdf(x=3,loc=0,scale=10)
```

```python
# percent point function
#it's inverse of cdf
print(stats.norm.ppf(q=0.025,loc=0,scale=10))
print(stats.norm.ppf(q=0.075,loc=0,scale=10))
```
